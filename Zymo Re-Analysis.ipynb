```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import numpy as np

# Sample data
data = [7.671288375, 8.610966242, 8.320876035, 8.371809964, 7.978342534, 8.468314289, 8.475164754, 8.50024916, 8.788699333, 8.683322903, 7.425655196, 8.006123075, 8.452245448, 7.874526371, 8.660971191, 7.677975794]

# Checking for normality using a normal probability plot
stats.probplot(data, plot=plt)
plt.show()

# Checking for normality using Shapiro-Wilk test
stat, p = stats.shapiro(data)
print('Shapiro-Wilk test:')
print('Statistics=%.3f, p=%.3f' % (stat, p))
alpha = 0.05
if p > alpha:
    print('Sample looks Gaussian (fail to reject H0)')
else:
    print('Sample does not look Gaussian (reject H0)')

# Checking for equal variances using Levene's test
group1 = [7.671288375, 8.610966242, 8.320876035, 8.371809964, 7.978342534]
group2 = [8.468314289, 8.475164754, 8.50024916, 8.788699333, 8.683322903]
stat, p = stats.levene(group1, group2)
print('Levene test:')
print('Statistics=%.3f, p=%.3f' % (stat, p))
alpha = 0.05
if p > alpha:
    print('Variances are equal (fail to reject H0)')
else:
    print('Variances are not equal (reject H0)')

# Checking for equal variances using Bartlett's test
stat, p = stats.bartlett(group1, group2)
print('Bartlett test:')
print('Statistics=%.3f, p=%.3f' % (stat, p))
alpha = 0.05
if p > alpha:
    print('Variances are equal (fail to reject H0)')
else:
    print('Variances are not equal (reject H0)')
```

```python
import numpy as np

def torgerson_scaling(distance_matrix, num_dimensions):
    # Step 1: Compute the centering matrix
    n = distance_matrix.shape[0]
    centering_matrix = np.identity(n) - (1/n)*np.ones((n,n))

    # Step 2: Compute the inner product matrix
    inner_product_matrix = -0.5 * np.dot(np.dot(centering_matrix, distance_matrix), centering_matrix)

    # Step 3: Compute the eigenvalues and eigenvectors of the inner product matrix
    eigenvalues, eigenvectors = np.linalg.eig(inner_product_matrix)

    # Step 4: Sort the eigenvalues and eigenvectors in descending order
    sorted_indices = np.argsort(eigenvalues)[::-1]
    eigenvalues = eigenvalues[sorted_indices]
    eigenvectors = eigenvectors[:,sorted_indices]

    # Step 5: Compute the matrix of transformed coordinates
    transformed_matrix = np.dot(eigenvectors[:,:num_dimensions], np.diag(np.sqrt(eigenvalues[:num_dimensions])))

    return transformed_matrix

# Sample distance matrix
distance_matrix = np.array([[0, 2, 4], [2, 0, 1], [4, 1, 0]])

# Perform Torgerson Scaling with 2 dimensions
transformed_matrix = torgerson_scaling(distance_matrix, 2)

# Print the transformed matrix
print(transformed_matrix)
```
