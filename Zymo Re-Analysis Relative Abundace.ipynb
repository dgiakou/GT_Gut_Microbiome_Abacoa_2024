```python
import numpy as np
import matplotlib.pyplot as plt


category_names = ['Strongly disagree', 'Disagree',
                  'Neither agree nor disagree', 'Agree', 'Strongly agree']
results = {
    'Question 1': [10, 15, 17, 32, 26],
    'Question 2': [26, 22, 29, 10, 13],
    'Question 3': [35, 37, 7, 2, 19],
    'Question 4': [32, 11, 9, 15, 33],
    'Question 5': [21, 29, 5, 5, 40],
    'Question 6': [8, 19, 5, 30, 38]
}


def survey(results, category_names):
    """
    Parameters
    ----------
    results : dict
        A mapping from question labels to a list of answers per category.
        It is assumed all lists contain the same number of entries and that
        it matches the length of *category_names*.
    category_names : list of str
        The category labels.
    """
    labels = list(results.keys())
    data = np.array(list(results.values()))
    data_cum = data.cumsum(axis=1)
    category_colors = plt.colormaps['RdYlGn'](
        np.linspace(0.15, 0.85, data.shape[1]))

    fig, ax = plt.subplots(figsize=(9.2, 5))
    ax.invert_yaxis()
    ax.xaxis.set_visible(False)
    ax.set_xlim(0, np.sum(data, axis=1).max())

    for i, (colname, color) in enumerate(zip(category_names, category_colors)):
        widths = data[:, i]
        starts = data_cum[:, i] - widths
        rects = ax.barh(labels, widths, left=starts, height=0.5,
                        label=colname, color=color)

        r, g, b, _ = color
        text_color = 'white' if r * g * b < 0.5 else 'darkgrey'
        ax.bar_label(rects, label_type='center', color=text_color)
    ax.legend(ncols=len(category_names), bbox_to_anchor=(0, 1),
              loc='lower left', fontsize='small')

    return fig, ax


survey(results, category_names)
plt.show()
```

```python
import requests
from bs4 import BeautifulSoup

# Send a request to the website
url = 'https://opencommons.uconn.edu/cgi/viewcontent.cgi?article=1510&context=law_review'
response = requests.get(url)

# Parse the HTML content with BeautifulSoup
soup = BeautifulSoup(response.content, 'html.parser')

# Find all the paragraphs in the document
paragraphs = soup.find_all('p')

# Initialize a counter for the number of sentences starting with "and"
and_count = 0

# Loop through each paragraph and count the number of sentences starting with "and"
for p in paragraphs:
    sentences = p.text.split('.')
    for s in sentences:
        if s.strip().startswith('The'):
            and_count += 1

# Print the result
print(f'The word "The" is used to begin a sentence {and_count} times in the document.')
```

```python
Pipeline modified from Qiime2 Moving Pictures Tutorial
Alyssa Demko

Project: Belize 2015 sediment culture-independent vs dependent comparison
06.13.2020

Only looking at 2020 controls with Vibrio spike in


#Activate Qiime2
###to activate: source activate qiime2-2020.2
###to exit environment: conda deactivate

#import files, note need absolute file path included in manifest doc
##files need to be unzipped

#Make sure you are in the proper directory (folder) and your files are there as well
pwd: /Users/alyssademko/Desktop/BZ15_2019/2020_ReAnalysis
#files:
pe-33-manifest.csv #file with path to raw reads for each sample
BZ15_metadata.tsv # include all metadata here for easier visualization/stats analysis later on
raw_reads/ #make sure they are unzipped!

#Import files
qiime tools import \
    --type 'SampleData[PairedEndSequencesWithQuality]' \
           --input-path pe-33-manifest-bzcontrols2.csv \
                        --output-path demux.qza \
                                      --input-format PairedEndFastqManifestPhred33

qiime demux summarize \
            --i-data demux.qza \
                     --o-visualization DEMUX_SEQ_SUMMARY_VIZ.qzv


#DADA2 denoising (takes a few hours- run overnight)
#this trims off the primers and trims the sequences (Reverse were more crappy which is why they were trimmed shorter), also removes chimeras (default is consensus method)
#decide metrics after looking at demux.qzv above

qiime dada2 denoise-paired \
            --i-demultiplexed-seqs demux.qza \
                                   --p-trunc-len-f 250 \
                                                   --p-trunc-len-r 155 \
                                                                   --p-trim-left-f 19 \
                                                                                   --p-trim-left-r 20 \
                                                                                                   --o-table table.qza \
                                                                                                             --o-representative-sequences rep-seqs.qza \
                                                                                                                                          --o-denoising-stats denoising-stats.qza



#visualize results from feature table
qiime feature-table summarize \
                    --i-table table.qza \
                              --o-visualization table.qzv \
                                                --m-sample-metadata-file BZ15_metadata.tsv

#visualize results summaries
qiime feature-table tabulate-seqs \
                    --i-data rep-seqs.qza \
                             --o-visualization rep-seqs.qzv

#visualize denoising stats
qiime metadata tabulate \
               --m-input-file denoising-stats.qza \
                              --o-visualization denoising-stats.qzv

#everything looks good from files above
#lowest depth after above step is AD069 @ 68,943 followed by AD072 @ 87,757
#will subsequently still need to remove chloroplast, singletons, etc


#Generate a tree for phylogenetic diversity analyses, start @ 1pm, finished 1:56pm
qiime phylogeny align-to-tree-mafft-fasttree \
                --i-sequences rep-seqs.qza \
                              --o-alignment aligned-rep-seqs.qza \
                                            --o-masked-alignment masked-aligned-rep-seqs.qza \
                                                                 --o-tree unrooted-tree.qza \
                                                                          --o-rooted-tree rooted-tree.qza


#Taxonomy
#make sure to use the classifier trained on the right scikit for this to work
#can take a few hours
qiime feature-classifier classify-sklearn \
                         --i-classifier silva-132-99-515-806-nb-classifier.qza \
                                        --i-reads rep-seqs.qza \
                                                  --o-classification silva-taxonomy.qza


qiime metadata tabulate \
               --m-input-file silva-taxonomy.qza \
                              --o-visualization silva-taxonomy.qzv


# need to get rid of chloroplasts! might as well do mitochondria too although not usually an issue
qiime taxa filter-table \
           --i-table table.qza \
                     --i-taxonomy silva-taxonomy.qza \
                                  --p-exclude mitochondria,chloroplast \
                                                           --o-filtered-table silva-filtered-table.qza

#visualize results from feature table
qiime feature-table summarize \
                    --i-table silva-filtered-table.qza \
                              --o-visualization silva-filtered-table.qzv \
                                                --m-sample-metadata-file BZ15_metadata.tsv



#Taxonomy Bar plots
#visualize filtered taxa bar plots
qiime taxa barplot \
           --i-table silva-filtered-table.qza \
                     --i-taxonomy silva-taxonomy.qza \
                                  --m-metadata-file BZ15_metadata.tsv \
                                                    --o-visualization silva-filtered-taxa-bar-plots.qzv




#####Filtering step specific to my experiment (threshold set at 70 reads)

qiime feature-table filter-features \
                    --i-table silva-filtered-table.qza \
                              --p-min-frequency 70 \
                                                --o-filtered-table silva-70-filtered-table.qza

#visualize
qiime feature-table summarize \
                    --i-table silva-70-filtered-table.qza \
                              --o-visualization silva-70-filtered-table.qzv \
                                                --m-sample-metadata-file BZ15_metadata.tsv

#Now lowest depth is AD069 @ 56,983 followed by AD067 @ 66,676

Taxonomy Bar plots
#visualize filtered taxa bar plots
qiime taxa barplot \
           --i-table silva-70-filtered-table.qza \
                     --i-taxonomy silva-taxonomy.qza \
                                  --m-metadata-file BZ15_metadata.tsv \
                                                    --o-visualization silva-70-filtered-taxa-bar-plots.qzv


#Results indicate that there are still ~50 genera that we only see in CD sampling! this includes streptomyces
#continue with analysis, first rarefaction


ALPHA RAREFACTION
#Alpha rarefaction makes table of diversity at different depths for your curve
#for first one, lets look at the unlimited diversity by doing max depth of 266,648
qiime diversity alpha-rarefaction \
                --i-table silva-70-filtered-table.qza \
                          --i-phylogeny rooted-tree.qza \
                                        --p-max-depth 266648 \
                                                      --m-metadata-file BZ15_metadata.tsv \
                                                                        --o-visualization alpha-rarefaction.qzv

#not enough steps, redo with >10 (default)
qiime diversity alpha-rarefaction \
                --i-table silva-70-filtered-table.qza \
                          --i-phylogeny rooted-tree.qza \
                                        --p-steps 20 \
                                                  --p-max-depth 266648 \
                                                                --m-metadata-file BZ15_metadata.tsv \
                                                                                  --o-visualization alpha-rarefaction-20steps.qzv



#### Lets Rarefy to 56,983 to include all samples and go from there...

qiime diversity alpha-rarefaction \
                --i-table silva-70-filtered-table.qza \
                          --i-phylogeny rooted-tree.qza \
                                        --p-max-depth 56983 \
                                                      --m-metadata-file BZ15_metadata.tsv \
                                                                        --o-visualization alpha-rarefaction-56k.qzv

#I think it looks okay in terms of being saturated


#now do diversity core metrices
qiime diversity core-metrics-phylogenetic \
                --i-phylogeny rooted-tree.qza \
                              --i-table silva-70-filtered-table.qza \
                                        --p-sampling-depth 56983 \
                                                           --m-metadata-file BZ15_metadata.tsv \
                                                                             --output-dir 56k/core-metrics-results-56k#%%
```

```python
import numpy as np
from scipy.linalg import solve
from sympy import symbols, Eq, integrate, sin, pi

# Solving a system of linear equations using SciPy
# 3x + 2y = 2
# x - 4y = 11

A = np.array([[3, 2], [1, -4]])
B = np.array([2, 11])
x, y = solve(A, B)
print("Solving linear equations with SciPy:")
print(f"x = {x:.2f}, y = {y:.2f}\n")

# Performing symbolic integration using SymPy
x = symbols('x')
expr = sin(2 * pi * x)
integral = integrate(expr, (x, 0, 1))
print("Symbolic integration with SymPy:")
print(f"Integral of sin(2*pi*x) from 0 to 1: {integral}")
```

```python
from sympy import symbols, Eq, solve, simplify_logic

# Define the symbols used in the logic equation
A, B, C = symbols("A B C")

# Define the logic equation
equation = Eq((A & ~B) | (B & C), True)

# Solve the equation for all combinations of A, B, and C
solutions = []

for a_val in [True, False]:
    for b_val in [True, False]:
        for c_val in [True, False]:
            if equation.subs({A: a_val, B: b_val, C: c_val}):
                solutions.append({A: a_val, B: b_val, C: c_val})

print("Solutions:")
for solution in solutions:
    print(solution)
```

```python
from ete3 import Tree, TreeStyle, NodeStyle, faces, TextFace

# Read the Newick-formatted string
newick_string = "(( (Bacteroides fragilis NCTC 9343_Phylum_Bacteriodetes:0.14775, Bifidobacterium longum subsp. longum JCM 1217_Phylum_Actinobacteria:0.13892) :0.01857, Helicobacter pylori strain ATCC 43504_Phylum_Proteobacteria:0.11945) :0.00449, Ruminococcus bromii strain ATCC 27255_Phylum_Firmicutes:0.11881, (Clostridium butyricum strain VPI3266_Phylum_Firmicutes:0.03359,Clostridium perfringens ATCC 13124_Phylum_Firmicutes:0.04132) :0.05316);"
t = Tree(newick_string)

# Define the layout function
def custom_layout(node):
    if node.is_leaf():
        # Create an italic face for the name
        name_face = TextFace(node.name, fsize=14, fgcolor="black", ftype="Arial", fstyle="italic")

        # Add the face to the leaf node
        faces.add_face_to_node(name_face, node, column=0, position="branch-right")

# Configure the tree style
ts = TreeStyle()
ts.show_leaf_name = False  # Set to False since we are using custom_layout function
ts.show_branch_length = True
ts.show_branch_support = True
ts.layout_fn = custom_layout  # Set the custom layout function

# Plot the tree
t.render("%%inline", w=800, units="px", tree_style=ts)
```

```python
Pipeline modified from Qiime2 Moving Pictures Tutorial
Alyssa Demko

Project: Belize 2015 sediment culture-independent vs dependent comparison
06.13.2020

Only looking at 2020 controls with Vibrio spike in


#Activate Qiime2
###to activate: source activate qiime2-2020.2
###to exit environment: conda deactivate

#import files, note need absolute file path included in manifest doc
##files need to be unzipped

#Make sure you are in the proper directory (folder) and your files are there as well
pwd: /Users/alyssademko/Desktop/BZ15_2019/2020_ReAnalysis
#files:
pe-33-manifest.csv #file with path to raw reads for each sample
BZ15_metadata.tsv # include all metadata here for easier visualization/stats analysis later on
raw_reads/ #make sure they are unzipped!

#Import files
qiime tools import \
    --type 'SampleData[PairedEndSequencesWithQuality]' \
           --input-path pe-33-manifest-bzcontrols2.csv \
                        --output-path demux.qza \
                                      --input-format PairedEndFastqManifestPhred33

qiime demux summarize \
            --i-data demux.qza \
                     --o-visualization DEMUX_SEQ_SUMMARY_VIZ.qzv


#DADA2 denoising (takes a few hours- run overnight)
#this trims off the primers and trims the sequences (Reverse were more crappy which is why they were trimmed shorter), also removes chimeras (default is consensus method)
#decide metrics after looking at demux.qzv above

qiime dada2 denoise-paired \
            --i-demultiplexed-seqs demux.qza \
                                   --p-trunc-len-f 250 \
                                                   --p-trunc-len-r 155 \
                                                                   --p-trim-left-f 19 \
                                                                                   --p-trim-left-r 20 \
                                                                                                   --o-table table.qza \
                                                                                                             --o-representative-sequences rep-seqs.qza \
                                                                                                                                          --o-denoising-stats denoising-stats.qza



#visualize results from feature table
qiime feature-table summarize \
                    --i-table table.qza \
                              --o-visualization table.qzv \
                                                --m-sample-metadata-file BZ15_metadata.tsv

#visualize results summaries
qiime feature-table tabulate-seqs \
                    --i-data rep-seqs.qza \
                             --o-visualization rep-seqs.qzv

#visualize denoising stats
qiime metadata tabulate \
               --m-input-file denoising-stats.qza \
                              --o-visualization denoising-stats.qzv

#everything looks good from files above
#lowest depth after above step is AD069 @ 68,943 followed by AD072 @ 87,757
#will subsequently still need to remove chloroplast, singletons, etc


#Generate a tree for phylogenetic diversity analyses, start @ 1pm, finished 1:56pm
qiime phylogeny align-to-tree-mafft-fasttree \
                --i-sequences rep-seqs.qza \
                              --o-alignment aligned-rep-seqs.qza \
                                            --o-masked-alignment masked-aligned-rep-seqs.qza \
                                                                 --o-tree unrooted-tree.qza \
                                                                          --o-rooted-tree rooted-tree.qza


#Taxonomy
#make sure to use the classifier trained on the right scikit for this to work
#can take a few hours
qiime feature-classifier classify-sklearn \
                         --i-classifier silva-132-99-515-806-nb-classifier.qza \
                                        --i-reads rep-seqs.qza \
                                                  --o-classification silva-taxonomy.qza


qiime metadata tabulate \
               --m-input-file silva-taxonomy.qza \
                              --o-visualization silva-taxonomy.qzv


# need to get rid of chloroplasts! might as well do mitochondria too although not usually an issue
qiime taxa filter-table \
           --i-table table.qza \
                     --i-taxonomy silva-taxonomy.qza \
                                  --p-exclude mitochondria,chloroplast \
                                                           --o-filtered-table silva-filtered-table.qza

#visualize results from feature table
qiime feature-table summarize \
                    --i-table silva-filtered-table.qza \
                              --o-visualization silva-filtered-table.qzv \
                                                --m-sample-metadata-file BZ15_metadata.tsv



#Taxonomy Bar plots
#visualize filtered taxa bar plots
qiime taxa barplot \
           --i-table silva-filtered-table.qza \
                     --i-taxonomy silva-taxonomy.qza \
                                  --m-metadata-file BZ15_metadata.tsv \
                                                    --o-visualization silva-filtered-taxa-bar-plots.qzv




#####Filtering step specific to my experiment (threshold set at 70 reads)

qiime feature-table filter-features \
                    --i-table silva-filtered-table.qza \
                              --p-min-frequency 70 \
                                                --o-filtered-table silva-70-filtered-table.qza

#visualize
qiime feature-table summarize \
                    --i-table silva-70-filtered-table.qza \
                              --o-visualization silva-70-filtered-table.qzv \
                                                --m-sample-metadata-file BZ15_metadata.tsv

#Now lowest depth is AD069 @ 56,983 followed by AD067 @ 66,676

Taxonomy Bar plots
#visualize filtered taxa bar plots
qiime taxa barplot \
           --i-table silva-70-filtered-table.qza \
                     --i-taxonomy silva-taxonomy.qza \
                                  --m-metadata-file BZ15_metadata.tsv \
                                                    --o-visualization silva-70-filtered-taxa-bar-plots.qzv


#Results indicate that there are still ~50 genera that we only see in CD sampling! this includes streptomyces
#continue with analysis, first rarefaction


ALPHA RAREFACTION
#Alpha rarefaction makes table of diversity at different depths for your curve
#for first one, lets look at the unlimited diversity by doing max depth of 266,648
qiime diversity alpha-rarefaction \
                --i-table silva-70-filtered-table.qza \
                          --i-phylogeny rooted-tree.qza \
                                        --p-max-depth 266648 \
                                                      --m-metadata-file BZ15_metadata.tsv \
                                                                        --o-visualization alpha-rarefaction.qzv

#not enough steps, redo with >10 (default)
qiime diversity alpha-rarefaction \
                --i-table silva-70-filtered-table.qza \
                          --i-phylogeny rooted-tree.qza \
                                        --p-steps 20 \
                                                  --p-max-depth 266648 \
                                                                --m-metadata-file BZ15_metadata.tsv \
                                                                                  --o-visualization alpha-rarefaction-20steps.qzv



#### Lets Rarefy to 56,983 to include all samples and go from there...

qiime diversity alpha-rarefaction \
                --i-table silva-70-filtered-table.qza \
                          --i-phylogeny rooted-tree.qza \
                                        --p-max-depth 56983 \
                                                      --m-metadata-file BZ15_metadata.tsv \
                                                                        --o-visualization alpha-rarefaction-56k.qzv

#I think it looks okay in terms of being saturated


#now do diversity core metrices
qiime diversity core-metrics-phylogenetic \
                --i-phylogeny rooted-tree.qza \
                              --i-table silva-70-filtered-table.qza \
                                        --p-sampling-depth 56983 \
                                                           --m-metadata-file BZ15_metadata.tsv \
                                                                             --output-dir 56k/core-metrics-results-56k
```

```python
from Bio import SeqIO

def parse_fna_file(file_path):
    smallest = float('inf')  # Initialize smallest to positive infinity
    largest = float('-inf')  # Initialize largest to negative infinity

    # Parse the .fna file with SeqIO.parse
    for record in SeqIO.parse(file_path, 'fasta'):
        # The description field usually contains the "size=#" attribute
        description = record.description

        # Find the "size=#" in the description
        size_index = description.find('size=')
        if size_index != -1:
            # Extract the size value
            size_value = int(description[size_index+5:].split()[0])
            # Update smallest and largest if necessary
            smallest = min(smallest, size_value)
            largest = max(largest, size_value)

    return smallest, largest

# Replace 'your_file.fna' with your actual file path
file_path = '/Users/dimitriosgiakoumas/Desktop/zr10770.16S_230509.zymo/00...AllSamples.Bac16Sv34/DADA2_ASV_Distribution/sv.seqs.fna'
smallest, largest = parse_fna_file(file_path)

print(f'Smallest size value: {smallest}')
print(f'Largest size value: {largest}')
```

```python
import biom
import pandas as pd

def convert_biom_to_csv(biom_file, csv_file):
    # Load the BIOM file
    table = biom.load_table(biom_file)

    # Convert to a Pandas DataFrame
    df = table.to_dataframe(dense=True)

    # Write the DataFrame to a CSV file
    df.to_csv(csv_file)

# Example usage
biom_file = '/Users/dimitriosgiakoumas/Downloads/ASV_Table.biom'  # Replace with your .biom file path
csv_file = '/Users/dimitriosgiakoumas/Downloads/ASV_Table.csv'  # Replace with your desired output path

convert_biom_to_csv(biom_file, csv_file)
```

```python
import plotly.graph_objects as go
from PIL import Image

# Sample data: taxonomic composition of multiple microbiome samples
microbiomes['Sample1'] = {
    'None;Other': 0.075742067,
    'Euryarchaeota': 0.049116324,
    'Actinobacteria': 0.004214359,
    'Bacteroidetes': 0.084114591,
    'Chlamydiae': 5.78339E-05,
    'Chloroflexi': 0.007700276,
    'Cyanobacteria': 0.005267594,
    'Elusimicrobia': 0.000758561,
    'Fibrobacteres': 0.010849173,
    'Firmicutes': 0.704400609,
    'Fusobacteria': 2.9331E-05,
    'Lentisphaerae': 0.006111405,
    'NA': 0.003288403,
    'Nitrospirae': 3.57752E-05,
    'Planctomycetes': 0.009601286,
    'Proteobacteria': 0.006894289,
    'Saccharibacteria': 1.51341E-05,
    'Spirochaetae': 0.001372058,
    'Synergistetes': 0.019067822,
    'Tenericutes': 0.004268445,
    'Verrucomicrobia': 0.007094664
}

# Set the file path to save the files
file_path = "/Users/dimitriosgiakoumas/Library/CloudStorage/OneDrive-FloridaAtlanticUniversity/"

# Set the DPI of the saved image
dpi = 300

# Set the size of the plot
width = 1400
height = 600

# Create a pie chart for each sample
for sample in microbiomes:
    # Get the taxonomic composition of the sample
    composition = microbiomes[sample]

    # Create a list of labels and values for the pie chart
    labels = []
    values = []
    for domain in composition:
        percentage = composition[domain] * 100
        labels.append('{}: {:.2f}%'.format(domain, percentage))
        values.append(composition[domain])

    # Create a pie chart using Plotly
    fig = go.Figure(data=[go.Pie(labels=labels, values=values)])
    fig.update_layout(title_text='Total Taxonomic Composition', legend=dict(orientation="h", x=-1, y=.8), width=width, height=height)

    # Save the plot as a PNG file with the specified DPI
    fig.write_image(file_path + "plot.png", scale=dpi/72)

    # Convert the PNG file to a TIFF file with the specified DPI
    with Image.open(file_path + "plot.png") as im:
        im.save(file_path + "plot.tiff", dpi=(dpi, dpi))
```
